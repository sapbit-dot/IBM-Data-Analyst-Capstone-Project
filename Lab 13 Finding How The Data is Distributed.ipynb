### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
print("Column Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nSummary Information:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
print("\nMissing values before cleaning:")
print(df.isnull().sum())

# Fill numeric columns with mean
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    df[col].fillna(df[col].mean(), inplace=True)

# Fill categorical columns with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Confirm that rows still exist
print(f"\nData shape after cleaning: {df.shape}")
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
top_n = 15
top_employment = df['Employment'].value_counts().head(top_n)

plt.figure(figsize=(10,5))
sns.barplot(x=top_employment.index, y=top_employment.values, palette="viridis")
plt.xticks(rotation=90)
plt.title(f'Top {top_n} Employment Categories')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

top_YearsCodePro = df['YearsCodePro'].value_counts().head(top_n)

plt.figure(figsize=(10,5))
sns.barplot(x=top_YearsCodePro.index, y=top_YearsCodePro.values, palette="viridis")
plt.xticks(rotation=90)
plt.title(f'Top {top_n} Years Code Pro Categories')
plt.xlabel('Years Code')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
job_sat_counts = df['JobSat'].value_counts(dropna=False)

plt.figure(figsize=(8, 5))
plt.pie(
    job_sat_counts, 
    labels=job_sat_counts.index, 
    autopct='%1.1f%%', 
    startangle=140,
)

plt.title('Job Satisfaction Distribution')
plt.tight_layout()
plt.show()


print('Since 55% of the rows were blank and filled in by median, the pie chart clearly shows the skewed data due to filling in the gaps. If we ignore this, we can see that scores 6-9 are the most common, so we can conclude that most respondents do feel satisfied at work.')
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
df_have = df['LanguageHaveWorkedWith'].str.split(';').explode().str.strip()
df_want = df['LanguageWantToWorkWith'].str.split(';').explode().str.strip()

count_have = df_have.value_counts()
count_want = df_want.value_counts()

lang_compare = pd.DataFrame({
    'HaveWorkedWith': count_have,
    'WantToWorkWith': count_want
}).fillna(0).astype(int)

lang_compare['Total'] = lang_compare['HaveWorkedWith'] + lang_compare['WantToWorkWith']
lang_compare = lang_compare.sort_values(by='Total', ascending=False).drop(columns='Total')

lang_compare.plot(kind='bar', figsize=(12,6), width=0.8)
plt.title('Comparison of Programming Languages: Worked With vs. Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
unique_countries = df['Country'].dropna().unique()
unique_countries = sorted(unique_countries)
print(unique_countries)
!pip install pycountry-convert
import pycountry_convert as pc

def country_to_region(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_map = {
            'AF': 'Africa',
            'NA': 'North America',
            'OC': 'Oceania',
            'AN': 'Antarctica',
            'AS': 'Asia',
            'EU': 'Europe',
            'SA': 'South America'
        }
        return continent_map.get(continent_code, 'Other')
    except:
        return 'Unknown'

df['Region'] = df['Country'].apply(country_to_region)

manual_mapping = {
    'Republic of Korea': 'Asia',
    'Democratic People\'s Republic of Korea': 'Asia',
    'Iran, Islamic Republic of...': 'Asia',
    'United States of America': 'North America',
    'United Kingdom of Great Britain and Northern Ireland': 'Europe',
    'Venezuela, Bolivarian Republic of...': 'South America',
    'Micronesia, Federated States of...': 'Oceania',
    'Congo, Republic of the...': 'Africa',
    'Democratic Republic of the Congo': 'Africa',
    'Palestine': 'Middle East',
    'Hong Kong (S.A.R.)': 'Asia',
    'Nomadic': 'Unknown'
}

df['Region'] = df.apply(
    lambda row: manual_mapping.get(row['Country'], row['Region']),
    axis=1
)

unmapped = df[df['Region'] == 'Unknown']['Country'].unique()
print(unmapped)

df.replace('Prefer not to say', pd.NA, inplace=True)

# Create pivot table showing counts of RemoteWork status by Region
heatmap_data = pd.crosstab(df['Region'], df['RemoteWork'])

# Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='Blues')

plt.title('Remote Work Trends by Region')
plt.xlabel('Remote Work Preference')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
!pip install scipy
from scipy.stats import pearsonr, spearmanr

df_corr = df[['JobSat', 'YearsCodePro']].copy()
df_corr = df_corr.dropna(subset=['JobSat', 'YearsCodePro'])
df_corr['YearsCodePro'] = df_corr['YearsCodePro'].replace('Less than 1 year', 0.5)
df_corr['YearsCodePro'] = df_corr['YearsCodePro'].replace('More than 50 years', 55)
df_corr['YearsCodePro'] = df_corr['YearsCodePro'].astype(float)


pearson_corr, pearson_p = pearsonr(df_corr['JobSat'], df_corr['YearsCodePro'])
spearman_corr, spearman_p = spearmanr(df_corr['JobSat'], df_corr['YearsCodePro'])

print(f"Pearson correlation: {pearson_corr:.3f} (p={pearson_p:.3f})")
print(f"Spearman correlation: {spearman_corr:.3f} (p={spearman_p:.3f})")

plt.figure(figsize=(7,5))
sns.regplot(
    x='YearsCodePro',
    y='JobSat',
    data=df_corr,
    scatter_kws={'alpha':0.4},
    line_kws={'color':'red'}
)
plt.title('Job Satisfaction vs. Professional Coding Years')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Job Satisfaction (1–5)')
plt.tight_layout()
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
top_employment = df['Employment'].value_counts().nlargest(20).index
ct_top = pd.crosstab(df.loc[df['Employment'].isin(top_employment), 'Employment'], df['EdLevel'])

ct_top.plot(kind='bar', stacked=True, figsize=(20,10), colormap='tab20')
plt.title('Top 20 Employment Status vs Education Level')
plt.xlabel('Employment Status')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=90, ha='right')
plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

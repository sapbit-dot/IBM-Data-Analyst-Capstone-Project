### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
age_mapping = {
    "Under 18 years old": 16,
    "18-24 years old": 21,
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years or older": 70
}

df["AgeGroup"] = df["Age"].map(age_mapping)
df_filtered = df.dropna(subset=["ConvertedCompYearly", "AgeGroup"])

# Group by AgeGroup and calculate median compensation
age_comp = df_filtered.groupby("AgeGroup")["ConvertedCompYearly"].median()

# Sort by AgeGroup (numeric)
age_comp = age_comp.sort_index()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(age_comp.index, age_comp.values, marker='o', linestyle='-', color='blue')

plt.title("Median Yearly Compensation by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Median ConvertedCompYearly ($)")
plt.grid(True)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python

df_filtered = df.dropna(subset=["ConvertedCompYearly", "AgeGroup"])

# Filter for ages 25-45
df_focus = df_filtered[(df_filtered["AgeGroup"] >= 25) & (df_filtered["AgeGroup"] <= 45)]

# Group by AgeGroup and calculate median compensation
age_comp_focus = df_focus.groupby("AgeGroup")["ConvertedCompYearly"].median().sort_index()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(age_comp_focus.index, age_comp_focus.values, marker='o', linestyle='-', color='green')

plt.title("Median Yearly Compensation for Ages 25 to 45")
plt.xlabel("Age Group")
plt.ylabel("Median ConvertedCompYearly ($)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
